#### subplot(row,column,plots)

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,4,1)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
colors=(['red','yellow','black','blue','green'])
size=([10,20,30,40,50])
plt.scatter(x,y,c=colors,s=size)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y,color='red')
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y,width=0.4)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y,alpha=0.2)
plt.show()
```
